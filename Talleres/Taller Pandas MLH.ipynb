4. Realice un gráfico de dispersión de las variables "pm2.5" y "TEMP" con título, labels apropiados y color rojo para los datos. Describa: ¿Qué se puede inferir de dicho gráfico?

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Diagrama de dispersión Temperatura vs. PM2.5', fontsize = 20)
ax.scatter(x = clima['TEMP'], y = clima['pm2.5'], color = 'red')
plt.xlabel("Temperatura", fontsize = 15)
plt.ylabel("Concentración de PM 2.5", fontsize = 15)
plt.show()
```

Del gráfico de dispersión se puede inferir que no existe correlación entre las variables pm2.5 y TEMP. Las variables son independientes entre sí.

5. Calcule la probabilidad de encontrar en la Base de datos el valor de pm2.5 encima de 120

La probabilidad es igual a:   Prob[evento] = Numero elementos evento / Espacio muestral

En donde el espacio muestral M = 43824 calculado de la siguiente manera:

```python
M= len(df)
M
```

El evento corresponde a las observaciones de pm2.5 mayores a 120.  El número de elementos del evento es igual a 12483 calculado de la siguiente manera:

```python
pm_mayor_120 = len(df[df['pm2.5'] > 120])
pm_mayor_120
```

Por lo tanto,  la probabilidad es igual a 12483 / 43824.

```python
print(pm_mayor_120 / M)
```
